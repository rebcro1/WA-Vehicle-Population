```python
#Look at the the missing value rows
df_cars_nan=df_cars[df_cars['State'].isnull() == True]
```

```python
df_cars_nan
```

```python
#Check for mixed types before dropping rows
for col in df_cars.columns.tolist():
    weird = (df_cars[[col]].applymap(type) != df_cars[[col]].iloc[0].apply(type)).any(axis=1)
    if len (df_cars[weird]) > 0:
        print(col)
```

```python
#Representing 0.5% of all rows, rows with no county/state will be removed
df_cars.dropna(subset = ['State'], inplace=True)
```

```python
df_cars.shape
```

```python
#Check dataset for full duplicates
df_cars_dup = df_cars[df_cars.duplicated()]
```

```python
#Drop unneccessary column - month ending date
df_cars_2 = df_cars.drop(columns = ['ME Date'])
```

```python
#Create consistent naming conventions and shorter fields
df_cars_2.rename(columns = {'Battery Electric Vehicles (BEVs)' : 'Battery_BEVs'}, inplace =True)
```

```python
df_cars_2.rename(columns = {'Plug-In Hybrid Electric Vehicles (PHEVs)' : 'Hybrid_(PHEVs)'}, inplace = True)
```

```python
df_cars_2.rename(columns = {'Electric Vehicle (EV) Total' : 'EV_Total'}, inplace = True)
```

```python
df_cars_2.rename(columns = {'Non-Electric Vehicle Total' : 'Non-Electric_Total'}, inplace = True)
```

```python
df_cars_2.rename(columns = {'Month Ending' : 'Month_ending'}, inplace = True)
```

```python
df_cars_2.rename(columns = {'Total Vehicles' : 'Total_Veh'}, inplace = True)
```

```python
df_cars_2.rename(columns = {'Percent Electric Vehicles' : '%_Electric'}, inplace = True)
```

```python
df_cars_2.rename(columns = {'Vehicle Primary Use' : 'Primary_use'}, inplace = True)
```
